```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("PySparkTest").getOrCreate()
print(spark.version)
```

```python
data = [1, 2, 3, 4, 5]
rdd = spark.sparkContext.parallelize(data)
print(rdd.collect())
```

```python
rdd_squared = rdd.map(lambda x: x * x)
print(rdd_squared.collect())
```

```python
rdd_filtered = rdd.filter(lambda x: x % 2 == 0)
print(rdd_filtered.collect())
```

```python
text_data = ["Hello world", "Hello PySpark", "PySpark is fun"]
rdd_text = spark.sparkContext.parallelize(text_data)
words = rdd_text.flatMap(lambda line: line.split(" "))
word_count = words.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
print(word_count.collect())
```

```python
data = [("Alice", 25), ("Bob", 30), ("Cathy", 22)]
rdd = spark.sparkContext.parallelize(data)
age_increased = rdd.map(lambda x: (x[0], x[1] + 5))
print(age_increased.collect())
```

```python
nums = list(range(0, 1001))
len(nums)
```

```python
# Import the necessary libraries
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder.appName("MyApp").getOrCreate()

# Get the SparkContext from the SparkSession
sc = spark.sparkContext

# Now you can use the sc variable
nums_rdd = sc.parallelize(nums)
nums_rdd
```

```python
nums_rdd.collect()
```

```python
nums_rdd.take(7)
```

```python
squared_nums_rdd = nums_rdd.map(lambda x: x**2)
squared_nums_rdd.take(5)
```

```python
len (str(546))
```

```python
pairs=squared_nums_rdd.map(lambda x: (x, len(str(x))))
pairs.take(25)
#
```

```python
even_digit_pairs=pairs.filter(lambda x: (x[1]%2)==0)
even_digit_pairs.take(25)
#
```

```python
flipped_pairs=even_digit_pairs.map(lambda x: (x[1], x[0]))
flipped_pairs.take(25)
#
```

```python
grouped_pairs=flipped_pairs.groupByKey()
grouped_pairs.take(25)
```

```python
grouped = flipped_pairs.groupByKey().mapValues(list)
grouped.take(25)
```

```python
grouped = flipped_pairs.groupByKey()
grouped.take(25)
```

```python
grouped = grouped.map(lambda x: (x[0], list(x[1])))
grouped.take(1)
```

```python
grouped = grouped.map(lambda x: (x[0], list(x[1])))
grouped.take(2)
```

```python
averaged = grouped.map(lambda x: (x[0], sum(x[1])/len(x[1])))
averaged.collect()
```

```python
avaraged = grouped.map(lambda x: (x[0], sum(x[1])/len(x[1])))
avaraged.take(25)
```
